```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head()
```

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python

# tamano del grafico
fig = plt.figure(figsize=(10, 5)) 

# graficar

# a) Face cream
plt.plot(
    df['month_number'], # eje x
    df['total_profit'], # eje y
    label='facecream lineal', # etiquetado
    color="black", # color
    linewidth=1 # tamano de la curva
)

plt.scatter(
    df['month_number'], # eje x
    df['total_profit']  , # eje y
    #s, # tamano de los puntos
    #c="g", # color
    alpha=0.9, # intensidad color
   label="facecream scatter" # etiquetdo fijando posicion
)



plt.legend() # agregar etiquetado
plt.xlabel('month_number') # nombre eje x
plt.ylabel('$total_profit') # nombre eje y
plt.show() # mostrar grafico
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
# grafico
fig = plt.figure(figsize=(12, 8))

barWidth = 0.4

plt.bar(df['month_number'], df['facecream'],color='blue', width=barWidth, edgecolor='white', label='facecream')
plt.bar(df['month_number']+barWidth, df['facewash'],color='red', width=barWidth, edgecolor='white', label='facewash')


#plt.xticks(y_pos, people)
plt.xlabel('Month')
plt.ylabel('Total unit')
plt.legend()
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
# Pre-format DataFrame
stats_df = df.drop(['month_number','total_units','total_profit'], axis=1) #filtra datos sacando estás estiquetas
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
plt.ylabel('unit')
plt.xlabel('product')
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
import pandas as pd
df2=df.sum()[1:7]
nombre=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
explode = [0, 0, 0, 0.09, 0, 0]
plt.pie(df2, labels=nombre, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Distribución % de Ventas Anuales por producto')
plt.show()
```
